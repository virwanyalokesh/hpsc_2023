```python
import math
import numpy as np
import matplotlib.pyplot as plt
#x=[1,1,3,5,3,2]
```

```python
def func_e(x, n):
    e_approx = 1
    temp=1
    for i in range(1,n+1):
        temp=temp*x/i
        e_approx +=temp
    return e_approx
#m=np.linspace(-1,1,1000);
```

```python
x = np.linspace(0,100,1000)
y1 = func_e(x,100)
y2 =np.exp(x)


```

```python
plt.figure()
plt.plot(x, y1,'r')
plt.xlabel('x1')
plt.ylabel('func_e(x)')


plt.figure()
plt.plot(x, y2,'k')
plt.xlabel('x1')
plt.ylabel('numpy.exp(x)')
```

```python
plt.plot(x, y2,'k',label="Imported function")
plt.plot(x, y1,'o',label="my defined function")
plt.legend()
```

```python
import timeit

time_1=timeit.timeit(lambda: func_e(1, 10));
time_2=timeit.timeit(lambda: np.exp(1));

print(f"The executation time for the defined function is...= {time_1}");
print(f"The executation time for inbuilt function(np.exp_) is = {time_2}");
print(f"The time difference between nmpy.exp() and defined exp. fn is = {time_1-time_2}")
```
